```python
#1.Importing the dependencies
import pandas as pd  
import numpy as np  
import seaborn as sns 
import matplotlib.pyplot as plt  
import warnings   # Importing the warnings module for managing warnings
%matplotlib inline
warnings.filterwarnings('ignore')  # Ignoring warnings to avoid cluttering the output
pd.options.display.max_columns = 99  # Setting the maximum number of displayed columns in pandas
```

```python
#2.Loading the dataset
df=pd.read_csv('turkiye-student-evaluation_generic.csv')
```

```python
#3.Principal Component Analysis(PCA)-->used to reduce the data attributes
x=df.iloc[:,5:33] #choosing the values from q1 to q28
from sklearn.decomposition import PCA
pca=PCA(n_components=2,random_state=42)
x_pca=pca.fit_transform(x)
```

```python
x_pca
```

```python
#how much info we retained from the dataset
pca.explained_variance_ratio_.cumsum()[1]
```

```python
#6.Model Training-->K-Means Clustering
from sklearn.cluster import KMeans
distortions=[] #for finding best model find the model with less distortions and try different clusters
cluster_range=range(1,6)
for i in cluster_range:
    model=KMeans(n_clusters=i, init='k-means++', random_state=42)
    model.fit(x_pca)
    distortions.append(model.inertia_)
```

```python
#elbow method
plt.plot(cluster_range,distortions,marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortions')
plt.show()
```

```python
model=KMeans(n_clusters=3, init='k-means++', random_state=42)
model.fit(x_pca)
y=model.predict(x_pca)
```

```python
plt.scatter(x_pca[y==0,0], x_pca[y==0,1], s=50, c='red', label='Cluster-1')
plt.scatter(x_pca[y==1,0], x_pca[y==1,1], s=50, c='yellow', label='Cluster-2')
plt.scatter(x_pca[y==2,0], x_pca[y==2,1], s=50, c='green', label='Cluster-3')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=100, c='blue', label='Centroids') #for cluster centroids
plt.title('Cluster Of Students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
